```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("FuelEfficiency.csv")
print(df.shape)
```

```python
df.head()
```

```python
sns.set(rc={'figure.figsize':{12,10}})
#Assuming data of Gas usage varies what you are in
ax = sns.lineplot(data = df, y = 'CombMPG', x = '#Gears')
```

```python
ax = sns.lineplot(data = df, x = '#Gears', y = 'CombMPG', hue = 'Cylinders', style = 'Cylinders', ci = False, markers = True)
```

```python
ax = sns.barplot(data = df, y = 'CombMPG', x = '#Gears')
```

```python
ax = sns.barplot(data = df, y = 'CombMPG', x = '#Gears', ci = False, hue = 'Cylinders')
```

```python
ax = sns.regplot(data = df, x = 'CombMPG', y = '#Gears')
```

```python
ax = sns.boxplot(data = df, y = 'CombMPG', x = '#Gears',hue = 'Cylinders',  orient = 'v')
```

```python
cor = df[['CombMPG', '#Gears', 'Cylinders']].corr()
cor
```

```python
col = ['CombMPG','#Gears', 'Cylinders']
ax = sns.heatmap(cor, annot = True, annot_kws = {'size' : 50}, xticklabels = col, yticklabels = col)
```

```python
x = df['CombMPG'].values
y = df['#Gears'].values

sns.distplot(x, hist = False, rug = True, color = 'green')
sns.distplot(y, hist = False, rug = True, color = 'blue')
```

```python
meanx = df['CombMPG'].mean()
meany = df['#Gears'].mean()

sns.distplot(x, hist = False, rug = True, color = 'green')
sns.distplot(y, hist = False, rug = True, color = 'blue')

#MeanPlot
plt.axvline(meanx, 0, 1, color = 'red')
plt.axvline(meany, 0, 1, color = 'orange')
```
